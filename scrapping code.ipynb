```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/Demographics_of_Nigeria'

# Add headers to mimic a real browser
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 '
                  '(KHTML, like Gecko) Chrome/115.0.0.0 Safari/537.36'
}

# sending requests with header
page = requests.get(url, headers=headers)

# parse HTML
soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup.prettify())
```

```python
soup.find('table')
```

```python
soup.find_all('table')[6]
```

```python
soup.find('table', class_ ="wikitable sortable", style="text-align: right;")
```

```python
table = soup.find_all('table')[6]
```

```python
print(table)
```

```python
world_titles = table.find_all('th')
```

```python
world_titles
```

```python
world_table_titles = [title.text.strip() for title in world_titles]

print(world_table_titles)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = world_table_titles)
df
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data[1:-1]:
  row_data = row.find_all('td')
  individual_row_data = [data.text.strip() for data in row_data]

  length = len(df)
  df.loc[length] = individual_row_data
```

```python
df
```

```python
df.to_csv(r'population.csv', index = False)
```
